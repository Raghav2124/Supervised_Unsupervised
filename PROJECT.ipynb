```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.compose import ColumnTransformer
```

```python
data=pd.read_csv("loan_data.csv")
```

```python
data.head()
```

```python
data.sample(10)
```

```python
data.drop(["Loan_ID","Loan_Amount_Term"],axis=1,inplace=True)
```

```python
data.info()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.isnull().mean()*100
```

```python
data['Dependents']=data['Dependents'].replace(["0","1","2","3+"],[0,1,2,3])
data['Dependents']=data['Dependents'].fillna(data['Dependents'].mode().iloc[0])
data['Gender'][data["Gender"].isnull()]=data['Gender'].dropna().sample(data["Gender"].isnull().sum())
data['Self_Employed']=data['Self_Employed'].fillna(data['Self_Employed'].mode().iloc[0])
data['Credit_History']=data['Credit_History'].fillna(data['Credit_History'].mode().iloc[0])
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lab= LabelEncoder()
data['Loan_Status'] = lab.fit_transform(data['Loan_Status'])
data['Gender'] = lab.fit_transform(data['Gender'])
data['Married'] = lab.fit_transform(data['Married'])
data['Self_Employed'] = lab.fit_transform(data['Self_Employed'])
data['Education'] = lab.fit_transform(data['Education'])

```

```python
data
```

```python
data["Property_Area"].nunique()

data
```

```python
data['Property_Area']=data['Property_Area'].replace(["Rural","Semiurban","Urban"],[0,1,2])
```

```python
data.isna().sum()
```

```python
x= data.drop(['Loan_Status'],axis=1).values
y= data['Loan_Status'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
c=DecisionTreeClassifier(max_depth=4)
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 2)
classifier.fit(x_train, y_train)
# x_train.isnull()
```

```python
y_pre=classifier.predict(x_test)
accuracy_score(y_test,y_pre)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(x_train, y_train)
```

```python
y_pr=forest.predict(x_test)
accuracy_score(y_test,y_pr)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clas = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
clas.fit(x_train, y_train)
y_p = clas.predict(x_test)
accuracy_score(y_test, y_p)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_ = gnb.predict(x_test)
accuracy_score(y_test, y_)
```
