```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
x=data.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    plt.title("The Elbow Function")
    kmeans=KMeans(n_clusters = i, init = 'k-means++',random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of Cluster")
plt.ylabel("Wcss")
plt.show()
```

```python
kmeans=KMeans(n_clusters = 5, init = 'k-means++',random_state=42)
c=kmeans.fit_predict(x) 
```

```python
print(c)
```

```python
plt.scatter(x[c==0,0],x[c==0,1],s=100,c="red",label="Cluster 1")
plt.scatter(x[c==1,0],x[c==1,1],s=100,c="blue",label="Cluster 2")
plt.scatter(x[c==2,0],x[c==2,1],s=100,c="green",label="Cluster 3")
plt.scatter(x[c==3,0],x[c==3,1],s=100,c="cyan",label="Cluster 4")
plt.scatter(x[c==4,0],x[c==4,1],s=100,c="magenta",label="Cluster 5")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c="yellow",label="Centroids")
plt.title("Clusters of customers")
plt.xlabel("Annual income(k$)")
plt.ylabel("Spending Score(1-100)")
plt.legend()
plt.show()
```
