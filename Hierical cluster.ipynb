```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
x=data.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
```

```python
den=sch.dendrogram(sch.linkage(x,method="ward"))
plt.title("Dendoragtam")
plt.xlabel("Number of Customers")
plt.ylabel("Euclidean distances")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h=AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage="ward")
h_ped=h.fit_predict(x)
```

```python
print(h_ped)
```

```python
plt.scatter(x[h_ped==0,0],x[h_ped==0,1],s=100,c="red",label="Cluster 1")
plt.scatter(x[h_ped==1,0],x[h_ped==1,1],s=100,c="blue",label="Cluster 2")
plt.scatter(x[h_ped==2,0],x[h_ped==2,1],s=100,c="green",label="Cluster 3")
plt.scatter(x[h_ped==3,0],x[h_ped==3,1],s=100,c="cyan",label="Cluster 4")
plt.scatter(x[h_ped==4,0],x[h_ped==4,1],s=100,c="magenta",label="Cluster 5")

plt.title("Clusters of customers")
plt.xlabel("Annual income(k$)")
plt.ylabel("Spending Score(1-100)")
plt.legend()
plt.show()
```
